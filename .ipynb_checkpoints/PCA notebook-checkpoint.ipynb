# Principal Component Analysis

## Using Scikit-learn built in dataset.

```python
# Imports
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
cancer = load_breast_cancer()
```

```python
# Looking into the components of data available.
cancer.keys()
```

```python
# Loading the data as a data frame.
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
```

```python
# Checking the basic information about the data.
df.info()
```

```python
# Checking the head of the data.
df.head()
```

```python
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df), columns=cancer.feature_names)
df.head()
```

```python
pca = PCA(n_components=2)
p_df = pd.DataFrame(pca.fit_transform(scaled_df))
```

```python
p_df.head()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(p_df[0], p_df[1], c = cancer.target, alpha=0.7)
plt.xlabel('1st PCA Component')
plt.ylabel('2nd PCA Component')
plt.title('PCA')
plt.show()
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer.feature_names)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_comp, cmap='autumn')
plt.show()
```
